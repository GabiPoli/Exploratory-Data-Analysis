```python
# Python libraries
from fpdf import FPDF
from datetime import datetime, timedelta
import os
from PIL import Image

WIDTH = 210
HEIGHT = 297

''' First Page '''

pdf = FPDF()
pdf.add_page()
pdf.image(r'Images Stock Market\Stock Market.png',0,0,WIDTH)


''' Second Page '''
pdf.add_page()
pdf.image(r'Images Stock Market\Stock Market 2.png',0,0,WIDTH)

pdf.set_font('Arial','', 10)
col1="";
pdf.multi_cell(100, 5, col1 + '\n\n\n\n\n', 0, 1);


pdf.set_font('Arial','', 10)
col1="In this project I will be looking at data from some Eletric Vehicle Stocks. I will use pandas to get stock information, visualize different aspects of it, and finally I will look at a few ways of analyzing the risk of a stock, based on its previous performance history. In addition to the EV companies I included AAPL so we can play around and figure out if there is any correlation. In December 2020, it was confirmed that Apple is working to launch a car, and right now, plans to release a vehicle in three to six years.\n\n";
pdf.multi_cell(95, 5, col1 + '\n', 0, 1);


''' Third Page Monte Carlo '''
pdf.add_page()
pdf.image(r'Images Stock Market\Stock Market 3.png',0,0,WIDTH)

pdf.set_font('Arial','', 10)
col1="";
pdf.multi_cell(95, 5, col1 + '\n\n\n\n\n', 0, 1);

pdf.set_font('Arial','', 10)
col1="One of the most common ways to estimate risk is the use of a Monte Carlo simulation (MCS). To calculate the value at risk (VaR) of a portfolio, we can run a Monte Carlo simulation that attempts to predict the worst likely loss for a portfolio given a confidence interval over a specified time horizon (we always need to specify two conditions for VaR: confidence and horizon).\n\n";
pdf.multi_cell(95, 5, col1, 0, 1);



''' Fourth Page Charts '''
pdf.add_page()
pdf.image(r'Images Stock Market\Stock Market 1.png',0,0,WIDTH)
#pdf.set_font('Arial', 'B', 16)
#pdf.cell(50, 30, 'Election Poll 2016 Analysis',ln=1)


pdf.image(r'Images Stock Market\Chart1.png', 35,10,0,130)
pdf.image(r'Images Stock Market\Chart2.png', 20,110,0,150)   



''' Fifth Page Charts '''
pdf.add_page()
pdf.image(r'Images Stock Market\Stock Market 1.png',0,0,WIDTH)


pdf.image(r'Images Stock Market\Chart3.png', 0,0,0,200)
pdf.image(r'Images Stock Market\Chart4.png', 0,100,0,200)   



''' Sixth Page Charts '''
pdf.add_page()
pdf.image(r'Images Stock Market\Stock Market 1.png',0,0,WIDTH)

pdf.image(r'Images Stock Market\Chart7.png', 30,0,0,130)
pdf.image(r'Images Stock Market\Chart8.png', 30,80,0,130)   
pdf.image(r'Images Stock Market\Chart9.png',30,150,0,130)


''' Eighth Page Conclusion '''
pdf.add_page()
pdf.image(r'Images Stock Market\Stock Market 4.png',0,0,WIDTH)

pdf.set_font('Arial','', 10)
col1="";
pdf.multi_cell(100, 5, col1 + '\n\n\n\n\n', 0, 1);

pdf.set_font('Arial', 'B', 16)
col1="Data Analysis:\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);

          
pdf.set_font('Arial','', 10,)
col1="On tthe chart 1 and 2 we were looking for some correlation. It is interesting to see the bigger correlation among chineses eletric vehicles.  If you compare the plot correlation between Tesla and Nio vs XPEV and NIO it is clear that chinese stocks have a stronger correlation.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);

pdf.set_font('Arial','', 10,)
col1="On the next charts we could also do a correlation plot, to get actual numerical values for the correlation between the stocks' daily return values. By comparing the closing prices, we see an interesting relationship between AAPL and TSLA for example.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);


pdf.set_font('Arial','', 10,)
col1="Moving Forward, on the risk analysis, there are many ways we can quantify risk, one of the most basic ways using the information we've gathered on daily percentage returns is by comparing the expected return with the standard deviation of the daily returns.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);


pdf.set_font('Arial','', 10,)
col1="Value at risk using the bootstrap method. For this method we will calculate the empirical quantiles from a histogram of daily returns.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);


pdf.set_font('Arial','', 10,)
col1="Using the the daily returns histogram for Apple stock, I can use quantile to get the risk value for the stock.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);

pdf.set_font('Arial','', 10,)
col1="Using the quantile formula we get the result of -0.032847823135718504. which means that, with 95% confidence, our worst daily loss will not exceed 3%. If we have a 1 million dollar investment, our one-day 5% VaR is 0.032 * 1,000,000 = $30,000.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);

pdf.set_font('Arial','', 10,)
col1="Value at Risk using the Monte Carlo method to run many trials with random market conditions, then I'll calculate portfolio losses for each trial. After this, I'll use the aggregation of all these simulations to establish how risky the stock is.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);

pdf.set_font('Arial','', 10,)
col1="Using the geometric Brownian motion, which is technically known as a Markov process. This means that the stock price follows a random walk and is consistent with (at the very least) the weak form of the efficient market hypothesis: past price information is already incorporated and the next price movement is conditionally independent of past price movements.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);

pdf.set_font('Arial','', 10,)
col1="This means that the past information on the price of a stock is independent of where the stock price will be in the future, basically meaning, you can't perfectly predict the future solely based on the previous price of a stock.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);

pdf.set_font('Arial','', 10,)
col1="The last two charts we have looked at the 1% empirical quantile of the final price distribution to estimate the Value at Risk for the Tesla stock, which looks to be $43.08 for every investment of 475.05 (the price of one inital tesla stock). \n\n";
pdf.multi_cell(180, 5, col1, 0, 1);

pdf.set_font('Arial','', 10,)
col1="This means that for every initial stock you purchase your putting about $43.08 at risk 99% of the time from our Monte Carlo Simulation.\n\n";
pdf.multi_cell(180, 5, col1, 0, 1);
    
pdf.output('Stock Market.pdf', 'F')
```
